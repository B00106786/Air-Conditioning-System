```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
temperature_range = np.arange(-5, 40, 1)  
```

```python
FREEZING = fuzz.trapmf(temperature_range,[-15, -10, 0, 5])  
COLD = fuzz.trapmf(temperature_range,[0, 5, 10, 15])  
NORMAL = fuzz.trapmf(temperature_range,[10, 15, 20, 25])
WARM = fuzz.trapmf(temperature_range,[20, 25, 30, 35])  
HOT = fuzz.trapmf(temperature_range,[30, 35, 40, 45])
```

```python
fig, ax = plt.subplots(nrows=1, figsize=(10, 5))
ax.plot(temperature_range, FREEZING, 'b', linewidth=1.5, label='FREEZING')  
ax.plot(temperature_range, COLD, 'g', linewidth=1.5, label='COLD')  
ax.plot(temperature_range, NORMAL, 'r', linewidth=1.5, label='NORMAL')
ax.plot(temperature_range, WARM, 'y', linewidth=1.5, label='WARM')  
ax.plot(temperature_range, HOT, 'k', linewidth=1.5, label='HOT') 
ax.set_title('TEMPERATURE')
ax.set_xlabel('Celcius')
ax.set_ylabel('Degrees of Membership')
ax.legend()
plt.savefig("temperature.jpg")
```
