```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
acCommand = np.arange(-10, 10, 1)  
```

```python
cool = fuzz.trapmf(acCommand,[-15, -10, -5, -2])  
maintain = fuzz.trimf(acCommand,[-3, 0, 3])  
heat = fuzz.trapmf(acCommand,[2, 5, 10, 15])
```

```python
fig, ax = plt.subplots(nrows=1, figsize=(10, 5))
ax.plot(acCommand, cool, 'b', linewidth=1.5, label='cool')  
ax.plot(acCommand, maintain, 'g', linewidth=1.5, label='maintain')  
ax.plot(acCommand, heat, 'r', linewidth=1.5, label='heat') 
ax.set_title('AC Command')
ax.set_xlabel('Command')
ax.set_ylabel('Degrees of Membership')
ax.legend()
plt.savefig("acCommand.jpg")
```
